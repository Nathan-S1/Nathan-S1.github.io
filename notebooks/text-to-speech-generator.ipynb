Create a Text to Speech Function

```python
from gtts import gTTS
from IPython.display import Audio, display

def text_to_speech(text, output_format="wav"):
    """
    Convert text to speech and save as an audio file using gTTS.

    Parameters:
    text (str): The text to be spoken.
    output_format (str): Audio file format ("wav" or "mp3").
    """

    # Use gTTS for audio files
    tts = gTTS(text=text, lang='en')

    # Save the audio file in the specified format
    if output_format == "wav":
        file_name = "audio.wav"
        tts.save(file_name)
        display(Audio(file_name, autoplay=True))
    elif output_format == "mp3":
        file_name = "audio.mp3"
        tts.save(file_name)
        display(Audio(file_name, autoplay=True))

    print(f"Audio saved as {file_name}")
```

Create Voice

```python
# User inputs
text = input("Enter what you the voice to say: ")
output_format = input("Enter the audio format (wav or mp3): ").lower()

# Run text-to-speech conversion
text_to_speech(text, output_format)
```
