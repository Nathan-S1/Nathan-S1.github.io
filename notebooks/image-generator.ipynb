Import Libraries

```python
from diffusers import StableDiffusionPipeline
import torch
from bs4 import BeautifulSoup
import requests
from PIL import Image
from io import BytesIO
from IPython.display import display
```

Scrape Google Images

```python
def scrape_google_images(query, num_images=10):
    query = query.replace(" ", "+")
    url = f"https://www.google.com/search?tbm=isch&q={query}&safe=off"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    image_elements = soup.find_all("img", limit=num_images + 1)[1:]  #Skip the first image

    image_urls = [img["src"] for img in image_elements if "src" in img.attrs]
    return image_urls

def display_images(image_urls, verbose=False):
    """
    Displays images fetched from URLs.
    Logs failed attempts without showing them in the output.
    """
    for i, url in enumerate(image_urls):
        try:
            response = requests.get(url)
            img = Image.open(BytesIO(response.content))
            if verbose:  #Show only if verbose is True
                print(f"Image {i + 1}: {url} (Loaded Successfully)")
            display(img)
        except Exception as e:
            if verbose:  #Log the error only if verbose is True
                print(f"Image {i + 1}: {url} (Failed - {e})")
```

Download and Extract Model
Load Pipeline from model directory

```python
model_name = "stabilityai/stable-diffusion-2-1"
pipe = StableDiffusionPipeline.from_pretrained(model_name, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

Generate Image

```python
#Define the prompt
prompt = input("Enter your image description prompt: ")

#Scrape Google Images based on the prompt
print("Fetching Google Images for reference...")
image_urls = scrape_google_images(prompt, num_images=20)

#Display the fetched images
print(f"Displaying the first {len(image_urls)} images:")
display_images(image_urls)

#Generate an image using Stable Diffusion
print("Generating the image using Stable Diffusion, please wait...")
negative_prompt = "blurry, low quality, distorted, text, watermark"
image = pipe(prompt, negative_prompt=negative_prompt, num_inference_steps=50, guidance_scale=7.5).images[0]

#Display and save the generated image
display(image)
output_path = "generated_image.png"
image.save(output_path)
print(f"Generated image saved as {output_path}")
```
